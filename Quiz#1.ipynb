```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("survey_results_public.csv")
```

```python
my_data = data[['Age', 'YearsCode', 'Country']]
print("Extracted data:")
print(my_data.head())
```

```python
indexed_data = data.set_index('Country')
print("\nIndexed data:")
print(indexed_data.head())
```

```python
filtered_data = data[(data['ResponseId'] < 10) & (data['Country'] == 'United States of America')]
print("\nFiltered data:")
print(filtered_data.head())
```

```python
sorted_data = data.sort_values(by=['LearnCodeCoursesCert', 'TechList'], ascending=[True, False])
print("\nSorted data:")
print(sorted_data.head())
```

```python
mean_value = data['CompTotal'].mean()
std_deviation = data['CompTotal'].std()
median_value = data['CompTotal'].median()
min_value = data['CompTotal'].min() 
max_value = data['CompTotal'].max() 

print("\nMean:", mean_value)
print("Standard Deviation:", std_deviation)
print("Median:", median_value)
print("Minimum Value:", min_value)
print("Maximum Value:", max_value)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(data['Age'], data['CompTotal'])
plt.xlabel('Ages')
plt.ylabel('CompTotals')
plt.title('Bar chart of categories')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Age'], data['CompTotal'], marker='o')
plt.xlabel('Ages')
plt.ylabel('CompTotal')
plt.title('Line chart of values over years')
plt.grid(True)
plt.show()
```
